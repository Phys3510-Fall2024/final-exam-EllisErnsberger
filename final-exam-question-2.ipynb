### Step 3 - 4: Functions, Tests, and Plotting $\omega$ vs. $Q_0$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.optimize import minimize
import ipywidgets as widgets

# Define potential energy function
def potential_energy(Q, K=1.0, D=2.4, Phi=1.0):
    return 0.5 * K * Q**2 - 0.25 * D * Q**4 + 0.1667 * Phi * Q**6

# Define the force
def force(Q, K=1.0, D=2.4, Phi=1.0):
    return -K * Q + D * Q**3 - Phi * Q**5

# Equation of motion
def equation_of_motion(t, y, K, D, Phi):
    Q, v = y
    a = force(Q, K, D, Phi)
    return [v, a]

# Solve for dynamics
def solve_dynamics(Q0, v0, K=1.0, D=2.4, Phi=1.0, t_span=(0, 50), t_eval=None):
    if t_eval is None:
        t_eval = np.linspace(*t_span, 1000)

    return solve_ivp(
        equation_of_motion, t_span, [Q0, v0], t_eval=t_eval, args=(K, D, Phi), method='RK45'
    )

# Extract period from dynamics
def extract_period(solution):
    Q = solution.y[0]
    t = solution.t

    # Find local maxima
    maxima_indices = (np.diff(np.sign(np.diff(Q))) < 0).nonzero()[0] + 1
    if len(maxima_indices) < 2:
        return np.nan  # Cannot determine period with fewer than two maxima

    # Calculate period
    periods = np.diff(t[maxima_indices])
    return np.mean(periods)

# Compute angular frequency
def angular_frequency(period):
    if np.isnan(period) or period == 0:
        return np.nan
    return 2 * np.pi / period

# Generate plot of frequency vs. amplitude
def plot_frequency_vs_amplitude(K=1.0, D=2.4, Phi=1.0):
    amplitudes = np.linspace(-1.7, 1.7, 50)
    frequencies = []

    for Q0 in amplitudes:
        sol = solve_dynamics(Q0, 0, K, D, Phi)
        period = extract_period(sol)
        freq = angular_frequency(period)
        frequencies.append(freq)

    plt.figure(figsize=(8, 6))
    plt.plot(amplitudes, frequencies, label="$\\omega$ vs. $Q_0$", color="blue")
    plt.axhline(y=np.sqrt(K), color="red", linestyle="--", label="Analytic $\\omega$ at $Q_0 = 0$")
    plt.xlabel("Initial Amplitude $Q_0$")
    plt.ylabel("Angular Frequency $\\omega$")
    plt.title("Frequency vs. Amplitude for Anharmonic Oscillator")
    plt.legend()
    plt.grid()
    plt.show()

plot_frequency_vs_amplitude()
```

### Step 5: Describing the Plot

The plot of angular frequency $\omega$ vs. initial displacement $Q_0$ conveys the fundamental anharmonic nature of the oscillatory system. For small values of $Q_0$ (near $Q_0 = 0$), the system acts approximately harmonic with a relatively consistent $\omega$ as a result of the quadratic term ($\frac{1}{2} K Q^2$) having a much larger impact on the oscillator than it has for larger values of $Q_0$, when the higher-order terms begin to dominate the system. At this scale, the effects of the fourth and sixth-order terms are negligible. At intermediate values of $Q_0$, the $- \frac{1}{4} D Q^4$ term reduces the restoring force, consequently decreasing $\omega$ due to the term's negative sign. Once the positive $\frac{1}{6} \phi Q^6$ term overcomes the effects of $- \frac{1}{4} D Q^4$, the restoring force begins to increase, resulting in a greater $\omega$. This nonlinear transition of term dominance appears to occur at the cusp near $Q_0 = \pm 0.7$.
