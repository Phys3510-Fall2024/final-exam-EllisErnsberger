```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Parameters
K1 = 1
K2 = 0.2
C = 0.5
M1 = M2 = 1

# Potential energy function
def potential_energy(Q1, Q2):
    return 0.5 * K1 * Q1**2 + 0.5 * K2 * Q2**2 + (1/3) * C * Q1 * Q2**2

# Equations of motion
def equations_of_motion(t, y):
    Q1, P1, Q2, P2 = y
    dQ1dt = P1 / M1
    dP1dt = -K1 * Q1 - (1/3) * C * Q2**2
    dQ2dt = P2 / M2
    dP2dt = -K2 * Q2 - C * Q1 * Q2
    return [dQ1dt, dP1dt, dQ2dt, dP2dt]

# Initial conditions
Q1_0 = 0
Q2_0 = 1
P1_0 = 0
P2_0 = 0
initial_conditions = [Q1_0, P1_0, Q2_0, P2_0]

# Time mesh
t_span = (0, 50)
t_eval = np.linspace(0, 50, 200)

# Solve the system
solution = solve_ivp(equations_of_motion, t_span, initial_conditions, t_eval=t_eval)
Q1, Q2 = solution.y[0], solution.y[2]

# Create a grid for contour plot
Q1_grid, Q2_grid = np.meshgrid(np.linspace(-1.5, 1.5, 100), np.linspace(-1.5, 1.5, 100))
U_grid = potential_energy(Q1_grid, Q2_grid)

# Plotting
plt.figure(figsize=(8, 6))

# Contour plot
filled_contour = plt.contourf(Q1_grid, Q2_grid, U_grid, levels=50, cmap='hot')
plt.contour(Q1_grid, Q2_grid, U_grid, levels=50, colors='white', linewidths=1)

# Trajectory plot
plt.plot(Q1, Q2, color='blue', label="Trajectory")

# Labels and limits
plt.xlabel("$Q_1$")
plt.ylabel("$Q_2$")
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.legend()
plt.title("Potential Energy Contours and Trajectory")
plt.colorbar(filled_contour, label="$\Delta U$ (Potential Energy)")
plt.grid()
plt.show()
```

### Dynamics of Initial Displacements

The fundamental differences between the effects of $Q_1$ and $Q_2$ on the system can be found in the details of the coupling term $\frac{1}{3} C Q_1 Q_2^2$ and the scalars $K_1$ and $K_2$ affecting $Q_1$ and $Q_2$, respectfully, representing the "stiffness" of each mode of the system. More specifically, $Q_1$ has a large stiffness constant ($K_1 = 1$) compared to $Q_2$'s stiffness constant ($K_2 = 0.2$). This means $Q_1$ experiences a stronger restoring force for some displacement than $Q_2$, thus resulting in $Q_1$ oscillating more quickly. Since the coupling term does not treat $Q_1$ and $Q_2$ symmetrically, the term produces a dependence on $Q_2$ which is quadratic when $Q_1$ is nonzero. Conversely, when $Q_2$ is nonzero, the coupling term is linearly dependent on $Q_1$. In other words, the quadratic dependence scenario results in large changes while the linear dependence scenario results in smaller changes. 

The physical interpretation of these differences between $Q_1$ and $Q_2$ can be explained by their respective oscillatory contributions to the system. With the given parameters, $Q_1$ contributes as the stronger oscillator because of its stronger stiffness constant and resistance to the coupling term. $Q_2$, on the other hand, contributes less to the system with a smaller stiffness constant, making it more easily affected by the coupling term. These asymmetric contributions indicate that $Q_1$ is the dominant oscillator and $Q_2$ has a smaller effect.
